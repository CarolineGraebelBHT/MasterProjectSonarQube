# Model Performances
The results obtained from evaluating the performance and quality of all models is visualised.

```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# laod data 

# data import
current_dir = os.getcwd()

# construct path to the project data folder
data_dir = os.path.join(current_dir, '..', '..', 'Data', 'Models','CodeSmellTags')

# load SonarQube measure data (without duplicates)
xgb = pd.read_csv(os.path.join(data_dir, 'XGBoost_Evaluation_Results.csv'), low_memory=False)
rf = pd.read_csv(os.path.join(data_dir, 'RandomForest_Evaluation_Results.csv'), low_memory=False)
lr = pd.read_csv(os.path.join(data_dir, 'LogisticRegression_Evaluation_Results.csv'), low_memory=False)
cnn = pd.read_csv(os.path.join(data_dir, 'CNN_Evaluation_Results.csv'), low_memory=False)
cc = pd.read_csv(os.path.join(data_dir, 'ClassifierChain_Evaluation_Results.csv'), low_memory=False)

data_dir = os.path.join(current_dir, '..', '..', 'Data', 'Models','CodeSmells')
quant = pd.read_csv(os.path.join(data_dir, 'model_performance_quantitypred.csv'), low_memory=False)
```

# Evaluation for Quantity Prediction

```python
# reconfigure MLR string so that it provides a better format for xticks
quant['Model'] = quant['Model'].replace('MLR with Ridge Regression', 'MLR with\nRidge Regression')

# create a figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharex=True)

# plot for MAE
axes[0].bar(quant['Model'], quant['MAE'], color='skyblue')
axes[0].set_title('Mean Absolute Error (MAE)')
axes[0].set_xlabel('Model')
axes[0].set_ylabel('MAE Value')
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# plot for MSE
axes[1].bar(quant['Model'], quant['MSE'], color='salmon')
axes[1].set_title('Mean Squared Error (MSE)')
axes[1].set_xlabel('Model')
axes[1].set_ylabel('MSE Value')
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

# plot for R²
axes[2].bar(quant['Model'], quant['R²'], color='lightgreen')
axes[2].set_title('R-squared ($R^2$)')
axes[2].set_xlabel('Model')
axes[2].set_ylabel('R² Value')
axes[2].grid(axis='y', linestyle='--', alpha=0.7)

plt.suptitle('Code Smell Quantity Prediction Models: Performance Comparison', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('model_performance_quantitypred.png')
```

# Evaluation for Code Smell Classification
The evaluations are presented split by approach.

## Binary Relevance

```python
# copy xgb df to keep original data
df_xgb = xgb
df_xgb['Model'] = 'XGBoost'
df_rf = rf
df_rf['Model'] = 'Random Forest'
df_lr = lr
df_lr['Model'] = 'Logistic Regression'

# calculate the mean scores over all tags for each of the three base models
mean_xgb = df_xgb.drop(['Tag', 'Model'], axis=1).mean().reset_index()
mean_xgb.columns = ['Metric', 'Mean Score']
mean_xgb['Model'] = 'XGBoost'

mean_rf = df_rf.drop(['Tag', 'Model'], axis=1).mean().reset_index()
mean_rf.columns = ['Metric', 'Mean Score']
mean_rf['Model'] = 'Random Forest'

mean_lr = df_lr.drop(['Tag', 'Model'], axis=1).mean().reset_index()
mean_lr.columns = ['Metric', 'Mean Score']
mean_lr['Model'] = 'Logistic Regression'

# combine the dataframes containing the mean scores
df_mean = pd.concat([mean_xgb, mean_rf, mean_lr], ignore_index=True)

# pivot the table to be in the right shape for the plot
df_wide = df_mean.pivot_table(index='Model', columns='Metric', values='Mean Score')
df_wide = df_wide.reset_index()
df_wide.columns.name = None

fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharex=True, sharey=True)

# plot for recall
axes[0].bar(df_wide['Model'], df_wide['Recall'], color='skyblue')
axes[0].set_title('Recall')
axes[0].set_xlabel('Model')
axes[0].set_ylabel('')
axes[0].set_ylim(0, 1)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# plot for precision
axes[1].bar(df_wide['Model'], df_wide['Precision'], color='salmon')
axes[1].set_title('Precision')
axes[1].set_xlabel('Model')
axes[1].set_ylabel('')
axes[1].set_ylim(0, 1)
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

# plot for f1-score
axes[2].bar(df_wide['Model'], df_wide['F1-Score'], color='lightgreen')
axes[2].set_title('F1-Score')
axes[2].set_xlabel('Model')
axes[2].set_ylabel('')
axes[2].set_ylim(0, 1)
axes[2].grid(axis='y', linestyle='--', alpha=0.7)

plt.suptitle('Mean Score for Code Smell Classification Models under a Binary Relevance Approach', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('model_performance_binaryrelevance.png')
```
